# VISUALIZATION

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
%matplotlib inline


```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[9,8,7,6,5,4,3,2,1]
```

```python
#plotting a line
plt.plot(x,y)
plt.title("MY LINE PLOT")
plt.xlabel('x-axis label')
plt.ylabel("y-axis label")
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()

```

```python
sns.lineplot(x=x,y=y)
plt.title("MY SNS LINEPLOT")
plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

```python
plt.bar(x,y)
plt.title("my bar plot")
plt.xlabel("x-axis label")
plt.ylabel("y-axis label")

plt.show()
```

```python
sns.barplot(x)
plt.title("my bar plot")
plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

```python
plt.hist(x)
plt.title("my histogram")

plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
           
    
```

```python


sns.histplot(x)
plt.title("my sns histplot")
plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
plt.show()
```

```python
x=[20,30,25,35]
labels=['A','B','C','D']
colors=['Green','Blue','Yellow','Orange']
explode=[0,0.1,0,0]
plt.pie(x,labels=labels,colors=colors,explode=explode,startangle=90)
plt.title("my pie chart")
plt.axis("equal")
plt.Circle(0,0)
plt.show()

```

```python
x=[1,2,3,4,5,]
y=[3,5,2,8,6]
plt.scatter(x,y)
plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.title("my scatter plot")
plt.legend()
plt.show()

```
